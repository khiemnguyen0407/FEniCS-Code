```python
from dolfin import *
import matplotlib.pyplot as plt
import numpy as np
import mshr
```

## Full-field solution

```python
wavelength = np.array([1.0/25, 1.0/25]);
wavenumber = 2*np.pi/wavelength
C = Expression("2.0 + sin(k1*x[0]) * cos(k2*x[1])", k1=wavenumber[0], k2=wavenumber[1], degree=2)
f = Expression("exp(-amplitude*(pow(x[0] - 0.5, 2) + pow(x[1] - 0.5, 2) ))", amplitude = 10, degree = 2)
# create a very fine mesh so that it the finite element solution can capture
# the fluctuation in the material parameter C
mesh = UnitSquareMesh(200, 200)
V_ff = FunctionSpace(mesh, "Lagrange", 2)

u = TrialFunction(V_ff)
v = TestFunction(V_ff)
a = dot(grad(v), C * grad(u)) * dx 
L = f * v * dx
# solve PDE for the full-field solution
u_ff = Function(V_ff)
solve(a == L, u_ff, DirichletBC(V_ff, Constant(0.0), (
    lambda x, on_boundary: (near(x[0], 0, DOLFIN_EPS_LARGE) or near(x[0], 1, DOLFIN_EPS_LARGE)) and on_boundary)))


# plot material parameter and solution
plt.figure(figsize=(8,6), dpi=80)
plt.subplot(2,2,1)
plt.colorbar(plot(project(C, V_ff), title="Material Parameter")); plt.grid()

plt.subplot(2,2,2)
plt.colorbar(plot(project(f, V_ff), title="External Source")); plt.grid()

plt.subplot(2,2,3)
plt.colorbar(plot(u_ff, title="Full-field solution")); plt.grid()
```

## Microscopic BVP and macroscopic elastic tensor

### Mesh and periodic constraint

```python
vertices = np.array([[-wavelength[0]/2, -wavelength[1]/2], 
                     [wavelength[0]/2, -wavelength[1]/2],
                     [wavelength[0]/2, wavelength[1]/2],
                     [-wavelength[0]/2, wavelength[1]/2]])
domain = mshr.Polygon([Point(vertices[0,:]), Point(vertices[1,:]), 
                       Point(vertices[2,:]), Point(vertices[3,:])])
mesh_RVE = mshr.generate_mesh(domain, 12)

# volume of the RVE
vol = np.prod(wavelength)
print("Volume of RVE =", vol)

# class used to define the periodic boundary map
class PeriodicBoundary(SubDomain):
    # Constructor
    def __init__(self, vertices, TOL=DOLFIN_EPS_LARGE):
        super().__init__()
        self.TOL = TOL
        self.vertices = vertices
        # vectors generating periodicity
        self.a = np.zeros((2, 2))
        self.a[0,:] = vertices[1,:] - vertices[0,:]
        self.a[1,:] = vertices[3,:] - vertices[0,:]
        
        # vectors normal to vectors a[0,:] and a[1,:]
        self.n = np.zeros((2,2))
        self.n[0,:] = [ -self.a[0,1], self.a[0,0] ]
        self.n[1,:] = [ -self.a[1,1], self.a[1,0] ]
        
        assert np.dot(self.n[0,:], self.a[0,:]) < self.TOL
        assert np.dot(self.n[1,:], self.a[1,:]) < self.TOL
    
    def inside(self, x, on_boundary):
        # return True if x is on the left or bottom boundaries
        # AND NOT on one of the bottom-right or top-left vertices
        
        top_left = np.allclose(x, self.vertices[3,:], self.TOL)
        # bottom_left = np.allclose(x, self.vertices[0,:], self.TOL)
        bottom_right = np.allclose(x, self.vertices[1,:], self.TOL)
        #top_left = near(x[0], self.vertices[3,0]) and near(x[1], self.vertices[3,1])
        #bottom_left = near(x[0], self.vertices[0,0]) and near(x[1], self.vertices[0,1])
        #bottom_right = near(x[0], self.vertices[1,0]) and near(x[1], self.vertices[1,1])
        
        x_vector = x - self.vertices[0,:]
        bottom = (np.abs(np.dot(self.n[0,:], x_vector)) < self.TOL)
        left = (np.abs(np.dot(self.n[1,:], x_vector)) < self.TOL)
        
        return (left or bottom) and \
            (not (top_left or bottom_right)) and on_boundary
    # Map right boundary (R) to the left boundary (L)
    def map(self, x, y):
        # top right node
        if np.allclose(x, self.vertices[2,:]):
            y[0] = x[0] - (self.a[0, 0] + self.a[1,0])
            y[1] = x[1] - (self.a[0, 1] + self.a[1,1])
        # right boundary
        elif np.abs(np.dot(self.n[1,:], x - self.vertices[1,:])) < self.TOL:
            y[0] = x[0] - self.a[0,0]
            y[1] = x[1] - self.a[0,1]
        # top boundary
        else:  
            y[0] = x[0] - self.a[1,0]
            y[1] = x[1] - self.a[1,1]
```

```python
# Material parameter
class K(UserExpression):
    def __init__(self, k1, k2, **kargs):
        super().__init__()
        self.k1 = k1
        self.k2 = k2
        
    def eval_cell(self, values, x, cell):
        values[0] = 2 + np.sin(self.k1*x[0]) * np.cos(self.k2*x[1])
        
    def value_shape(self):
        return ()
    
kappa = K(wavenumber[0], wavenumber[1])

# kappa = Expression("2.0 + sin(k1*x[0]) * cos(k2*x[1])", 
#                    k1=wavenumber[0], k2=wavenumber[1], degree=2)

# Kinematics and kinetics variables
def epsilon(u):
    return grad(u)

def sigma(u, eps_):
    return kappa * (grad(u) + eps_)
```

### Fluctuation field: solution by removing rigid-body mode

```python
ElemU = FiniteElement("CG", mesh_RVE.ufl_cell(), 2)
ElemL = FiniteElement("R", mesh_RVE.ufl_cell(), 0)
W = FunctionSpace(mesh_RVE, MixedElement([ElemU, ElemL]), 
                 constrained_domain=PeriodicBoundary(vertices))

u, c = TrialFunctions(W)   # the fluctuation field
v, d = TestFunctions(W)

MacroGrad = Constant(np.array([1.0, 0]))

a = dot(grad(v), kappa*grad(u))*dx + c*v*dx + u*d*dx
L = -dot(grad(v), kappa * MacroGrad) *dx
# F = dot(epsilon(v), kappa * (grad(u) + MacroGrad))*dx + c*v*dx + u*d*dx 
# a, L = lhs(F), rhs(F)

w = Function(W)
solve(a == L, w, [])

u_fluc, lagrange_mul = w.split()

plt.figure(figsize=(14, 7), dpi=60)
plt.subplot(2,3,1)
plt.colorbar(plot(project(kappa, FunctionSpace(mesh_RVE, 'CG', 1)), 
     title="RVE material parameter")); plt.grid(); 
plt.subplot(2,3,3)
plt.colorbar(plot(u_fluc, title="fluctuation field")); plt.grid()
plt.subplot(2,3,4)
plt.colorbar(plot(u_fluc.dx(0), title="$u_x$")); plt.grid()
plt.subplot(2,3,6)
plt.colorbar(plot(u_fluc.dx(1), title="$u_y$")); plt.grid()
```

Now, we compute the macroscopic tangent $\overline{\mathbf{C}}$. This tangent is a second-order tensor and its components can be computed by solving only two microscopic boundary value problems. In fact, at the macroscopic level we know that the resultant constitutive laww is linear and given by
\begin{equation}
    \overline{\mathbf{q}} = \overline{\mathbf{C}}\cdot \overline{\mathbf{p}}
\end{equation}
Therefore, if we choose $\overline{\mathbf{p}_{i}}$ to be the unit basis vectors in $\mathbb{R}^2$, then we can derive $\overline{\mathbf{q}}_{i}$ as the $i^\text{th}$-column of $\overline{\mathbf{C}}$. Indeed, let us conosider $\overline{\mathbf{p}}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$. Then, it implies that
\begin{equation}
\overline{q}_i = \sum\limits_{j}\overline{C}_{ij} \overline{p}_j = C_{i\,:}
\end{equation}

```python
I = np.eye(2)
C_Macro = np.zeros((2, 2))

w = Function(W)
for irun in range(2):
    MacroStrain = Constant(I[:,irun])
       
    F = dot(epsilon(v), sigma(u, MacroStrain))*dx + c*v*dx + u*d*dx 
    a, L = lhs(F), rhs(F)
    
    solve(a == L, w, [])
    u_fluc, lagrange_mul = w.split()
    
    C_Macro[:, irun] = np.array([assemble(sigma(u_fluc, MacroStrain)[i]*dx)\
                     for i in range(2)]) / vol

print("C_Macro ="); print(C_Macro)
```

### Fluctuation field in terms of gradient variable

```python
p_order = 3;
ElemP = VectorElement("CG", mesh_RVE.ufl_cell(), p_order)
ElemL1 = FiniteElement("CG", mesh_RVE.ufl_cell(), p_order-1)
ElemL2 = VectorElement("R", mesh_RVE.ufl_cell(), 0)

W_Mixed = FunctionSpace(mesh_RVE,
                        MixedElement([ElemP, ElemL1, ElemL2]),
                        constrained_domain=PeriodicBoundary(vertices))

(p, r, s) = TrialFunctions(W_Mixed)
(q, c, d) = TestFunctions(W_Mixed)

# Define bilinear form a, and linear form L
a = dot(q,  kappa*p)*dx + c * (p[0].dx(1) - p[1].dx(0))*dx \
    + r * (q[0].dx(1) - q[1].dx(0))*dx + dot(s, q)*dx + dot(p, d)*dx
L = -dot(q, kappa * MacroGrad)*dx
# Solve for solution and then split it into separate components
# corresponding to each component of the mixed function space
sol = Function(W_Mixed)
solve(a == L, sol, [])
p_fluc, mu_, lambda_ = sol.split()

# Plot the fluctation gradient fields
plt.figure(figsize=(12.5, 3.0), dpi=60)
plt.subplot(1,3,1)
plt.colorbar(plot(p_fluc[0], title="$p_1 = u_{x}$")); plt.grid()

plt.subplot(1,3,3)
plt.colorbar(plot(p_fluc[1], title="$p_2 = u_{x}$")); plt.grid()
```

```python
I = np.eye(2)
C_Macro = np.zeros((2, 2))

w = Function(W)
for irun in range(2):
    MacroGrad = Constant(I[:,irun])
       
    # Define bilinear form a, and linear form L
    a = dot(q,  kappa*p)*dx + c * (p[0].dx(1) - p[1].dx(0))*dx \
            + r * (q[0].dx(1) - q[1].dx(0))*dx + dot(s, q)*dx + dot(p, d)*dx
    L = -dot(q, kappa * MacroGrad)*dx
    
    solve(a == L, sol, [])
    p_fluc, mu_, lambda_ = sol.split()
    
    C_Macro[:, irun] = np.array([assemble(kappa*(MacroGrad[i] + p_fluc[i])*dx)\
                     for i in range(2)]) / vol

print("C_Macro =");
print(C_Macro)
```

### Solve the homogenized boundary value problem

```python
mesh_Macro = UnitSquareMesh(25, 25, "crossed")
V_Macro = FunctionSpace(mesh_Macro, "CG", 1)

bc = DirichletBC(V_Macro, 0, lambda x, on_boundary: (
    near(x[0], 0, DOLFIN_EPS_LARGE) or near(x[0], 1, DOLFIN_EPS_LARGE)) \
    and on_boundary)

u = TrialFunction(V_Macro)
v = TestFunction(V_Macro)
a = dot(grad(v), Constant(C_Macro) * grad(u)) * dx 
L = f * v * dx
# solve PDE
u_homo = Function(V_Macro)
solve(a == L, u_homo, bc)
```

```python
plt.figure(figsize=(10, 8), dpi=80)

plt.subplot(2,2,1)
plt.colorbar(plot(u_homo, title="homogenized solution")); plt.grid()

plt.subplot(2,2,2)
plt.colorbar(plot(u_ff, title="full-field solution")); plt.grid()

plt.subplot(2,2,3)
plt.colorbar(plot(project(u_homo, V_ff) - u_ff,
             title="$u_{\mathrm{homo}} - u_{\mathrm{full}}$"))
```
