```python
from dolfin import *
import matplotlib.pyplot as plt
```

```python
# Unit Interval Mesh

mesh = UnitIntervalMesh(10)
print("Plotting a UnitIntervalMesh")
plt.figure()
plot(mesh, title="Unit interval")
```

```python
# Unit Square Mesh

plt.figure(figsize=(7, 7), dpi=80)
plt.subplot(2, 2, 1)
plot(UnitSquareMesh(10, 10, "right"), title="Diagonals to the right")

plt.subplot(2, 2, 2)
plot(UnitSquareMesh(10, 10, "left"), title="Diagonals to the left")

plt.subplot(2, 2, 3)
plot(UnitSquareMesh(10, 10, "crossed"), title="Crossed diagonals")


plt.subplot(2, 2, 4)
plot(UnitSquareMesh(10, 10, "right/left"), title="right/left")
```
